```python
#read downloaded files
path = '/Users/hemingyi/Dropbox/New_York_SVI/image'         # 替换路径
dowonloaded_file = os.listdir(path)              
fopen = open('/Users/hemingyi/Dropbox/New_York_SVI/name.txt', 'w') # 替换路径
downloaded = set()
for d in dowonloaded_file:                       
    a,b = d.split('.',1)
    string = a + '\n'  
    fopen.write(string)
    downloaded.add(a)
fopen.close()
print ('done')
```
